Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
cleaned_students = []    #creating a new array to store cleaned data
```

```python
# Loop through the raw_students for each student to clean data

for student in raw_students:
    #clean name
    name = student["name"].strip().title()

    # Convert roll to "int"
    roll = int(student["roll"])

    # Convert marks_str to list of integers
    marks = [int(m.strip()) for m in student["marks_str"].split(",")]

    #Validate name to contain only alphabetic characters
    valid = all(w.isalpha() for w in name.split())
    if valid:
        print(f"{name} : ✓ Valid name")
    else:
        print(f"{name} : ✗ Invalid name")

    # Store cleaned student
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })    
```

```python
# Print a formatted profile card for each cleaned student using f-string
for student in cleaned_students:
    print("================================")
    print(f"Student : {student['name']}")
    print(f"Roll No : {student['roll']}")
    print(f"Marks   : {student['marks']}")
    print("================================")    
```

```python
# After processing all students, print the name in ALL CAPS and lowercase for the student with roll number 103.
for student in cleaned_students:
    if student['roll'] == 103:
        print("\nRoll 103:")
        print(student['name'].upper())
        print(student['name'].lower())
```

Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print(f"Report Card for {student_name}:\n")
# Grade Mapping
for i in range(len(subjects)):
    mark = marks[i]

    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"  

    print(f"{subjects[i]}: {mark} -> Grade: {grade}")      
```

```python
# Total marks (sum of all subjects)
print(f"Total Marks: {sum(marks)}")

# Average marks (rounded to 2 decimal places)
print(f"Average marks: {(round(sum(marks)/len(marks)),2)}")

# Highest scoring subject (name + marks)
print(f"Highest scoring subject: {subjects[marks.index(max(marks))]} ({max(marks)})")

# Lowest scoring subject (name + marks)
print(f"Lowest scoring subject: {subjects[marks.index(min(marks))]} ({min(marks)})")
```

```python
new_subjects_added = 0

while True:
    subject = input("Enter subject name (or type 'done' to stop): ").strip()
    
    if subject.lower() == "done":
        break
    
    marks_input = input(f"Enter marks for {subject} (0–100): ").strip()
    
    # Validate marks input
    try:
        mark = float(marks_input)
        
        if mark < 0 or mark > 100:
            print("✗ Invalid marks! Must be between 0 and 100. Skipping entry.\n")
            continue
        
    except ValueError:
        print("✗ Invalid input! Please enter a numeric value. Skipping entry.\n")
        continue
    
    # If valid, add to lists
    subjects.append(subject)
    marks.append(mark)
    new_subjects_added += 1
    print(f"✓ Added {subject} with marks {mark}\n")

# Final calculations
total = sum(marks)
average = round(total / len(marks), 2)

print("\n--- Summary ---")
print(f"New subjects added: {new_subjects_added}")
print(f"Updated Average Marks: {average}")
```

Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
pass_count = 0
fail_count = 0
total_avg_sum = 0
topper_name = ""
topper_avg = 0

print("Name              | Average | Status")
print("----------------------------------------")

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    
    # Counting pass/fail
    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1
    
    # Tracking topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name
    
    total_avg_sum += avg
    
    print(f"{name:<17} | {avg:>7.2f} | {status}")

# Class average
class_avg = round(total_avg_sum / len(class_data), 2)

print("\nSummary:")
print(f"Passed: {pass_count}, Failed: {fail_count}")
print(f"Topper: {topper_name} ({topper_avg:.2f})")
print(f"Class Average: {class_avg:.2f}")

```

Task 4 — String Manipulation Utility

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# Strip leading and trailing whitespace from essay. Store this as clean_essay and use it for all steps below.
clean_essay = essay.strip()
print(f"Step 1 - Clean Essay: {clean_essay}")
```

```python
# Convert clean_essay to Title Case and print it.
title_case = clean_essay.title()
print(f"Step 2 - Title Case: {title_case}")
```

```python
# Count how many times the word "python" appears in clean_essay (case-insensitive). Print the count.
count_python = clean_essay.count("python")
print(f"Step 3 - Count of 'python': {count_python}")
```

```python
# Replace every occurrence of "python" in clean_essay with "Python 🐍" and print the result.
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Step 4 - Replaced Essay: ")
print(replaced_essay)
print()
```

```python
# Split clean_essay into individual sentences by splitting on ". " (a period followed by a space). Print the resulting list.
sentences = clean_essay.split(". ")
print("Step 5 - Sentence List:")
print(sentences)
print()
```

```python
# Print each sentence on a new line, numbered starting from 1. Add a "." at the end of each sentence if it does not already end with one.
print("Step 6 - Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
